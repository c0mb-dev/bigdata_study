```python
#리스트를 생성하고 데이터 프레임으로 생성
import csv 
import pandas as pd

b_date = []
a_temperature = []
l_temperature = []
h_temperature = []


f = open('seoul.csv', 'r', encoding='cp949') 
data = csv.reader(f, delimiter=',') 
for row in data :
    if row[0][5:] == '04-26':
        b_date.append(row[0])
        a_temperature.append(row[1])
        l_temperature.append(row[2])
        h_temperature.append(row[3])
        
f.close() 

b_calendar = pd.DataFrame({"날짜": b_date, "평균 기온": a_temperature,\
                           "최저 기온": l_temperature, "최고 기온": h_temperature})
b_calendar #출력
```

```python
import numpy as np #데이터 분석 라이브러리
#https://www.cubrid.org/manual/ko/9.1.0/sql/datatype.html
#https://stackoverflow.com/questions/31917964/python-numpy-cannot-convert-datetime64ns-to-datetime64d-to-use-with-numba

b_calendar.info()
b_calendar['날짜'] = b_calendar['날짜'].astype('datetime64[ns]')
b_calendar['최고 기온'] = b_calendar['최고 기온'].astype(float)
b_calendar['최저 기온'] = b_calendar['최저 기온'].astype(float)
b_calendar['평균 기온'] = b_calendar['평균 기온'].astype(float)
print("******************데이터 변환******************")
b_calendar.info()
```

```python
b_calendar.sort_values(by='최저 기온').head()
#2009년 4월 26일이 제일 추웠음
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

%matplotlib inline
#%matplotlib inline, notebook을 실행한 브라우저에서 바로 그림을 볼 수 있게 해줌 
#[출처]https://korbillgates.tistory.com/85

font_path = 'C:/Windows/Fonts/a아시아헤드2.ttf'
fontprop = fm.FontProperties(fname=font_path, size=18)
#폰트 설정


plt.figure(figsize=(20,10))
plt.plot(b_calendar['날짜'], b_calendar['평균 기온'], 'rs--') #rs--, 스타일
plt.title('날짜별 기온', fontproperties=fontprop)
plt.ylabel('기온', fontproperties=fontprop)
plt.xlabel('날짜', fontproperties=fontprop)
plt.grid()
plt.show()

#날짜의 데이터가 XXXX-04-26 로 되어 있어서 그래프가 밀려나는 현상
```
